```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
import pandas as pd
```

```python
array_4x4 = np.random.exponential(scale=1.0, size=(4, 4))
print("4x4 array from exponential distribution:\n", array_4x4)
```

```python
exp_data = np.random.exponential(scale=1.0, size=100000)
uniform_data = np.random.rand(100000)
normal_data = np.random.randn(100000)

plt.figure(figsize=(10, 6))
plt.hist(exp_data, density=True, bins=100, histtype="step", color="red", label="Exponential")
plt.hist(uniform_data, density=True, bins=100, histtype="step", color="blue", label="Uniform")
plt.hist(normal_data, density=True, bins=100, histtype="step", color="green", label="Normal")

plt.legend(loc="upper right")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('$Z = X^2 + Y^2$')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file_path = '/mnt/data/Au_nanoparticle_dataset.csv'
df = pd.read_csv(file_path)

print("Columns in the dataset:\n", df.columns)
columns_to_keep = ['N_total_actual', 'N_bulk_actual', 'N_surface_actual', 'R_avg_actual']
filtered_df = df[columns_to_keep]
print("First 20 samples of the new dataframe:\n", filtered_df.head(20))
stats = filtered_df.describe().loc[['mean', 'std', '25%', '50%', '75%']]
print("Statistics for each feature:\n", stats)
plt.figure(figsize=(20, 5))
for i, column in enumerate(columns_to_keep):
    plt.subplot(1, 4, i+1)
    sns.histplot(filtered_df[column], kde=True)
    plt.title(f'Histogram of {column}')
plt.tight_layout()
plt.show()
sns.pairplot(filtered_df)
plt.show()

g = sns.PairGrid(filtered_df)
g.map_upper(sns.histplot)  
g.map_diag(sns.histplot, kde=True) 
g.map_lower(sns.kdeplot)
plt.show()
```
